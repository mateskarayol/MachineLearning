Clustering
    

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

# Loading data and labels
data = np.load('data.npy')
label = np.load('label.npy')

# Setting color and marker for each label
color_map = {0 : 'm', 1 : 'y', 2 : 'c'}
color = []

for key in label:
    color.append(color_map[key])

# Plotting the datapoints in relevant colors
plt.scatter(data[:,0],
            data[:,1],
            s=5,
            marker='.',
            c=color )

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

In this question, data is printed in different colors based on labels.

```python

# Running K-means algorithm via scikitlearn library
kmeans = KMeans(n_clusters=3, 
                init='random',
                n_init=10)
kmeans.fit_predict(data)

# ------------------------------------------
# QUESTION 3
# ------------------------------------------

# Plotting the found clusters and cluster centers
print('----------------------------------')
print('CLUSTER CENTERS are shown with *')
print('----------------------------------')
print(kmeans.cluster_centers_)
print('----------------------------------')

# Assignning different colors for each label calculated in k-means 
color = []
for key in kmeans.labels_:
    color.append(color_map[key])
    
# Plotting data
plt.scatter(data[:,0],
            data[:,1],
            s=5,
            marker='.',
            c=color )

# Plotting cluster centers
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1], 
            s=100, 
            marker='*', 
            c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```


In 2nd and 3rd questions,  scikit learn k-means algorithm is used to assign data to different clusters.
*  n_clusters parameter is assigned as 3 because there are three clusters as it is seen in previous question
  
*  n_init parameter is assigned as 10 so that kmeans algorithm is ran by 10 times and choose min Sum of Squared Errors
  
*  init parameter is assigned as random so that first selection for centroids is done randomly

Cluster centers is reachable via cluster_centers_ attribute.
Cluster labels is reachable via cluster_labels_ attribute.
Data is plotted in different colors based on cluster and cluster centers are plotted as star.
